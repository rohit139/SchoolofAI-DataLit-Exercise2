```
import pandas as pd
```

```
url ='https://media.githubusercontent.com/media/WillKoehrsen/Data-Analysis/master/univariate_dist/data/formatted_flights.csv'
flight_data=pd.read_csv(url)

```

```
flight_data.head(10)
```

```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# matplotlib Histogram
plt.hist(flight_data['arr_delay'],bins=int(180/5),color='blue',edgecolor = 'black')

plt.title('Histogram of Arrival Delays')
plt.xlabel('Delay (min)')
plt.ylabel('Flights')
```

```
# seaborn histograms
sns.distplot(flight_data['arr_delay'],bins=int(180/5),color='blue')

plt.title('Histogram of Arrival Delays')
plt.xlabel('Delay (min)')
plt.ylabel('Flights')
```

```
sns.distplot(flight_data['arr_delay'],bins=int(180/5),color='blue',hist=True,kde=False,hist_kws={'edgecolor':'red'})

plt.title('Histogram of Arrival Delays')
plt.xlabel('Delay (min)')
plt.ylabel('Flights')
```

```
airlines_master=flight_data.name.unique()
airlines_master
```

```
airlines1=airlines_master[:5]
airlines1
```

```
airlines2=airlines_master[5:10]
airlines2
```

```
airlines3=airlines_master[10:]
airlines3
```

```
for airline in airlines1:
  
  #generate the subset for the specific airine
  subset=flight_data[flight_data['name']==airline]
  
  sns.distplot(subset['arr_delay'],hist=False,kde=True,kde_kws={'linewidth':3},label=airline)
  
plt.legend(prop={'size': 16}, title = 'Airline')
plt.title('Density Plot with Multiple Airlines')
plt.xlabel('Delay (min)')
plt.ylabel('Density')
```

```
for airline in airlines2:
  
  #generate the subset for the specific airine
  subset=flight_data[flight_data['name']==airline]
  
  sns.distplot(subset['arr_delay'],hist=False,kde=True,kde_kws={'linewidth':3},label=airline)
  
plt.legend(prop={'size': 16}, title = 'Airline')
plt.title('Density Plot with Multiple Airlines')
plt.xlabel('Delay (min)')
plt.ylabel('Density')
```

```
for airline in airlines3:
  
  #generate the subset for the specific airine
  subset=flight_data[flight_data['name']==airline]
  
  sns.distplot(subset['arr_delay'],hist=False,kde=True,kde_kws={'linewidth':3},label=airline)
  
plt.legend(prop={'size': 16}, title = 'Airline')
plt.title('Density Plot with Multiple Airlines')
plt.xlabel('Delay (min)')
plt.ylabel('Density')
```

As per the above 3 charts we see the 3 best pefroming airlines as as follow : 1.Alaska, 2. United and 3. Envoy Air.

So the next step will be to compare among these thee to see whic is the best performing among them.


```
airlinesList=['United Air Lines Inc.','Envoy Air','Alaska Airlines Inc.','American Airlines Inc.']
```

```
for airline in airlinesList:
  
  #generate the subset for the specific airine
  subset=flight_data[flight_data['name']==airline]
  
  sns.distplot(subset['arr_delay'],hist=False,kde=True,
               kde_kws={'shade': True,'linewidth':3},
               label=airline)
  
plt.legend(prop={'size': 16}, title = 'Airline')
plt.title('Density Plot with Multiple Airlines')
plt.xlabel('Delay (min)')
plt.ylabel('Density')
```

```
airlinesList=['United Air Lines Inc.','Alaska Airlines Inc.','American Airlines Inc.']

for airline in airlinesList:
  
  #generate the subset for the specific airine
  subset=flight_data[flight_data['name']==airline]
  
  sns.distplot(subset['arr_delay'],hist=False,kde=True,
               kde_kws={'shade': True,'linewidth':3},
               label=airline)
  
plt.legend(prop={'size': 16}, title = 'Airline')
plt.title('Density Plot with Multiple Airlines')
plt.xlabel('Delay (min)')
plt.ylabel('Density')
plt.xlim(-5,None)
```
